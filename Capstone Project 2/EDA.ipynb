**Because differences between values of features are very high to observe on plot (like the area_mean feature's max value is 2500 and smoothness_mean features' max 0.16340.), so we need standardisation before visualization, feature selection**

```python
data.shape
```

```python
#Determine if ANY Value in a data is Missing
data.isnull().values.any()  
```

```python
data.drop(['Unnamed: 32'], axis=1).isnull().values.any()  
```

```python
data.diagnosis.unique()
```

```python
#Find duplicate rows in data
data[data.duplicated()]
```

```python
source = data.drop(['id','Unnamed: 32','diagnosis'],axis=1)
source.head()
```

```python
target = data.diagnosis
target.unique()
```

```python
target.value_counts()
```

```python
ax = sns.countplot(target,label="Count")       
B, M = target.value_counts()
```

```python
# standardization
source_s = (source - source.mean()) / (source.std()) 
source_s.head()
#source_s.describe()
```

```python
data = pd.concat([target,source_s.iloc[:,0:10]],axis=1)
data.head()
```

```python
data = pd.melt(data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
data.head()
```

**In Box Plot, we plot features in 3 group and each group includes 10 features to observe better.**

```python
# box plots are also useful in terms of seeing outliers

plt.figure(figsize=(10,10))
sns.boxplot(x="features", y="value", hue="diagnosis", data=data)
plt.xticks(rotation=90)
```

```python
data = pd.concat([target,source_s.iloc[:,10:20]],axis=1)
data = pd.melt(data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')

plt.figure(figsize=(10,10))
sns.boxplot(x="features", y="value", hue="diagnosis", data=data)
plt.xticks(rotation=90)
```

```python
data = pd.concat([target,source_s.iloc[:,20:31]],axis=1)
data = pd.melt(data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')

plt.figure(figsize=(10,10))
sns.boxplot(x="features", y="value", hue="diagnosis", data=data)
plt.xticks(rotation=90)
```

**variable of concavity_worst and concave point_worst looks like similar**

**Also it seems radius_worst, perimeter_worst and area_worst are correlated as it can be seen pair plot. We use these for feature selection.**

```python
df = source_s.loc[:,['radius_worst','perimeter_worst','area_worst']]
sns.pairplot(df)
```

**Now in swarm Plot, we plot features in 3 group and each group includes 10 features to observe better.**

```python
data = pd.concat([target,source_s.iloc[:,0:10]],axis=1)
data = pd.melt(data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')

plt.figure(figsize=(10,10))
sns.swarmplot(x="features", y="value", hue="diagnosis", data=data)

plt.xticks(rotation=90)
```

**I think perimeter_mean and area_mean looks like malignant and benign are seprated not totaly but mostly.**
